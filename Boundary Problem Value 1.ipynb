```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f(u,x):
    y,z=u
    f1,f2=z,1-y
    return[f1,f2]
x=np.linspace(0,np.pi/2.0,50)
y0=0.0
z1,z2=1.0,-2.0
u=[y0,z1]
soln=odeint(f,u,x)
w1=soln[:,0][-1]
plt.plot(x,soln[:,0], '--', label='Guess value z=1.0')
#u=[y0,z2]
#soln=odeint(f,u,x)
#w2=soln[:,0][-1]
#plt.plot(x,soln[:,0], '--', label='Guess value z=-2.0')
tol=0.0001
for i in range(1000):
    u=[y0,z2]
    sol=odeint(f,u,x)
    w2=soln[:,0][-1]
    if abs(w1-w2)<tol:
        break
    z3=z2+(z2-z1)/(w2-w1)*(y0-w2)
    print ("The interpolated value is",z3)
    z1,z2=z2,z3
    w1=w2

plt.plot(x,soln[:,0],'o',label="Calculated")
plt.plot(x,1-np.sin(x)-np.cos(x),'x',label="Exact solution")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Boundary Value Problem')
plt.grid(True)
plt.legend()
plt.savefig('Boundary Value Problem2.pdf')
plt.show()
```
